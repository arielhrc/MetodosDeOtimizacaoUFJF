```python
from plotly import graph_objects as go
import numpy as np

#aqui começamos declaramos os vetores com os intervalos especificados no exercicio
#por padrao o np.linspace() usa 50 amostras no intervalo, no caso aqui de -6 a 6, dando um salto de 0.24 por valor de eixo_x 
eixo_x = np.linspace(-6,6)
eixo_y = np.linspace(-6,6)
#aqui declaro a funçao dada no exercicio
eixo_z = lambda x,y: x**2 + y**2 + 25*(np.sin(x)**2 + np.sin(y)**2)
#aqui eu crio uma matriz de resultados 50x50 pra facilitar na hora de plotar depois
z_final = np.zeros((len(eixo_x),len(eixo_y)))
#aqui vamos preencher essa matriz utilizando os dois vetores como indices da matriz
for i in range(len(eixo_x)):
  for j in range(len(eixo_y)):
    #z_final recebe funçao z de x e y
    z_final[i][j]= eixo_z(eixo_x[i], eixo_y[j])
#aqui finalmente plotamos o grafico. eu tinha como opçao o Scatter3d, o Mesh3d e o Surface, mas esses dois primeiros eu nao consegui plotar
ovo = go.Figure(data=go.Surface(x = eixo_x, y = eixo_y, z = z_final))
ovo.show()
```
